### Q12
Students in two classes:
```python
class_A = {'Aman','Riya','Karan','Neha'}

```

````python
# write your code here
class_A = {'Aman','Riya','Karan','Neha'}
class_B = {'Neha','Simran','Aman'}
total = class_A.union(class_B)
print("Total number of students :",len(total))
````
