## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
# write your code here
n = int(input())
for i in range(2, n - 1):
  if n % i == 0:
    print("Not a prime number.")
    break
else :
  print("Prime number.")
```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
# write your code here
n = int(input())

for i in range(2, n + 1) :
  flag = False
  for j in range(2, i):
    if i % j == 0 :
      flag = True
      break
  if flag == False:
    print(i)





```

## Question 19 (Hard)

**Title:** Number Pattern

### Problem Statement:
Take N as input and print a right-angled number pattern using nested loops.

Example:
1
12
123

### Approach:
Use an outer loop for rows and an inner loop for printing numbers.


```python
# write your code here

var=int(input())

for i in range(1, var + 1):
  for j in range(1, i + 1):
    print(j, end="")
  print()



```
